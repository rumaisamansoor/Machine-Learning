## Finding Similar avg age as Pakistan 

```
Pak=data.loc['Pakistan',:]
from sklearn.metrics.pairwise import cosine_similarity
sim_mat = cosine_similarity([Pak], data)[0]  #similarity with each countru
sort = sorted(sim_mat)[-6:]
sort
```

```
indxs=[]
for sor in sort:   
    indx = np.where(sim_mat == sor)[0][0]
    indxs.append(indx)
data.iloc[indxs]
```

```
pak_sim=data.iloc[indxs]
pak_sim
```

```
fig=plt.figure(figsize=(20, 8))
for c in range(len(pak_sim)):
    row = pak_sim.loc[pak_sim.index[c], :]
    plt.plot(row, colors[c%len(colors)], label=pak_sim.index[c], linewidth=2, marker='o')
    plt.xticks(rotation=90)
    plt.legend(loc='upper left',prop={'size': 16})
    plt.xlabel('Year')
    plt.ylabel('Life Exp')                  #all of these show very similar trend as Pakistan
plt.show()
```
